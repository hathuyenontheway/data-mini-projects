# **Analyze Weather in Ho Chi Minh from 2020 to 2025**
**First step**: request weather dataset from **open-meteo**
1. Using requests lib to get dataset
2. Save csv

```python
import requests
import pandas as pd
import json

url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": 10.8231,       
    "longitude": 106.6297,
    "start_date": "2020-01-01",
    "end_date": "2025-12-31",
    "daily": [
        "temperature_2m_max",
        "temperature_2m_min",
        "precipitation_sum",
        "windspeed_10m_max",
        "precipitation_sum",
        "rain_sum"
    ],
    "timezone": "Asia/Ho_Chi_Minh"
}

response = requests.get(url, params=params)
data = response.json()
df = pd.DataFrame(data["daily"])
df.to_csv("hcmc_weather_2020_2025.csv")
```
